```python
# 외계인 기타 연주 

import sys

input = sys.stdin.readline
n,P = map(int, input().split())
ans = 0

stk = [[] for _ in range(7)]
for _ in range(n):
    line, p = map(int, input().split())
    
    # 1) p보다 큰 프렛들 모두 떼기
    while stk[line] and stk[line][-1] > p:
        stk[line].pop()
        ans += 1 #손 떼기

    # 2) 이제 꼭대기가 p면 이미 누른 상태 → 아무 것도 안 함
    if stk[line] and stk[line][-1] == p:
        continue
        

    # 3) 스택이 비었거나 꼭대기가 p보다 작으면 p를 새로 누름
    if not stk[line] or stk[line][-1] < p:
        stk[line].append(p)
        ans += 1  # 손 누르기 동작

print(ans)
```
